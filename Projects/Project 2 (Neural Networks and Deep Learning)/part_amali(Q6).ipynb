```python
import random

allcustomers = 2000


def funcserviceTime():
    rand_num = random.random()
    if rand_num <= 0.35:
        return 12
    elif 0.35 < rand_num <= 0.61:
        return 14
    elif 0.61 < rand_num <= 0.77:
        return 16
    elif 0.77 < rand_num <= 0.91:
        return 18
    elif 0.91 <= rand_num < 1:
        return 20





def func_for_interarrival():
    rand_num = random.random()
    if rand_num <= 0.3:
        return 1
    elif 0.3 < rand_num <= 0.81:
        return 2
    elif 0.81 < rand_num <= 1:
        return 3

st = [funcserviceTime()]
it = [0]

class Person:
    def __init__(self, arrival_time, line_num=1, position=1):
        self.arrival_time = arrival_time
        self.begin_time = 10 ** 6
        self.end_time = 10 ** 6
        self.line_num = line_num
        self.position = position

    def calculate_waiting_time(self):
        return self.begin_time - self.arrival_time
allcustomerss = [Person(0, position=1)]



ii = 0
i2 = 0

for i in range(1, allcustomers, 1):
    it.append(func_for_interarrival())
    if i % 5 == 0:
        st.append(funcserviceTime())
    arrival_time = allcustomerss[i - 1].arrival_time + it[i]
    allcustomerss.append(Person(arrival_time=arrival_time, position=i % 5 + 1))

time = [0] * 6
total_time = 0

length = [0] * (allcustomerss[-1].end_time + 1)


for i in range(1, len(length)):
    length[i] += length[i - 1]


endTime = 0

while ii < allcustomers:

    if allcustomerss[ii + 4].arrival_time <= endTime:
        for i in range(5):
            allcustomerss[ii + i].begin_time = endTime
            allcustomerss[ii + i].end_time = allcustomerss[ii + i].begin_time + st[i2]
    else:
        for i in range(5):
            allcustomerss[ii + i].begin_time = allcustomerss[ii + 4].arrival_time
            allcustomerss[ii + i].end_time = allcustomerss[ii + i].begin_time + st[i2]

    endTime = allcustomerss[ii].begin_time + st[i2]
    i2 += 1
    ii += 5


for customer in allcustomerss:
    length[customer.arrival_time] += 1
    length[customer.begin_time + 1] -= 1
    time[customer.position] += customer.begin_time - customer.arrival_time
    total_time += customer.end_time - customer.arrival_time

# /////////////




wtm=sum(time) / allcustomers
print("waitingTime mean= : ",wtm )

sm=total_time / allcustomers
print("system mean",sm )

wp=sum(st) / allcustomerss[-1].end_time * 100
print("work percentage=",wp )

ll= sum(length) / allcustomerss[-1].end_time
print("line length=", ll)

for i in range(0,5):
    temp= time[i + 1] / (allcustomers / 5)
    print("each group mean=", i + 1, "is : ",temp)


# ////


allcustomerss = [Person(0, position=1)]
for i in range(1, allcustomers, 1):
    arrival_time = allcustomerss[i - 1].arrival_time + it[i]
    allcustomerss.append(Person(arrival_time=arrival_time, position=i % 5 + 1))

ii = 0
i2 = 0

work_time = [0] * 2

endtime1 = 0
endtime2 = 0


time = [0] * 6
total_time = 0

while ii < allcustomers:
    if endtime1 <= endtime2:
        for i in range(5):
            allcustomerss[ii + i].begin_time = max(endtime1, allcustomerss[ii + 4].arrival_time)
            allcustomerss[ii + i].end_time = allcustomerss[ii + i].begin_time + st[i2]
    else:
        if allcustomerss[ii + 4].arrival_time < endtime1:
            for i in range(5):
                allcustomerss[ii + i].begin_time = endtime2
                allcustomerss[ii].line_num = 2
                allcustomerss[ii + i].end_time = allcustomerss[ii + i].begin_time + st[i2]
        elif endtime2 <= allcustomerss[ii + 4].arrival_time < endtime1:
            for i in range(5):
                allcustomerss[ii + i].begin_time = allcustomerss[ii + 4].arrival_time
                allcustomerss[ii].line_num = 2
                allcustomerss[ii + i].end_time = allcustomerss[ii + i].begin_time + st[i2]
        else:
            for i in range(5):
                allcustomerss[ii + i].begin_time = allcustomerss[ii + 4].arrival_time
                allcustomerss[ii + i].end_time = allcustomerss[ii + i].begin_time + st[i2]

    if allcustomerss[ii].line_num != 2:
        endtime1 = allcustomerss[ii].begin_time + st[i2]
    else:
        endtime2 = allcustomerss[ii].begin_time + st[i2]

    work_time[allcustomerss[ii].line_num - 1] += st[i2]

    ii += 5

    i2 += 1



length2 = [0] * (allcustomerss[-1].end_time + 1)

for customer in allcustomerss:
    length2[customer.arrival_time] += 1
    length2[customer.begin_time + 1] -= 1
    time[customer.position] += customer.begin_time - customer.arrival_time
    total_time += customer.end_time - customer.arrival_time

for i in range(1, len(length2)):
    length2[i] += length2[i - 1]


wtm2=sum(time) / allcustomers
print("waitingTime mean 2= ",wtm2 )

wp2= work_time[0] / allcustomerss[-1].end_time * 100
print("work percentage=",wp2 )

ssm2= total_time / allcustomers
print("system2 mean=",ssm2)

ll2= sum(length2) / allcustomerss[-1].end_time
print("line length2= ",ll2)

for i in range(0,5):
    temp2=time[i + 1] / (allcustomers / 5)
    print("each group mean=", i + 1, "is : ", temp)


```
